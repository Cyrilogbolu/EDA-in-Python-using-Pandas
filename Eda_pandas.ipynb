```python
df = df[['coaster_name', 
        #'Length', 'Speed', 
           'Location', 'Status', 
           'Opening date',
       #'Type', 
           'Manufacturer',
           #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
      # 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      # 'Single rider line available', 'Restraint Style',
      # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
           #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
           #'height_value', 'height_unit',
           'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
# dropping a column

df = df.drop(["Opening date"], axis = 1)
```

```python
#changing the data type of date
df["opening_date_clean"] = pd.to_datetime(df["opening_date_clean"])
```

```python
df = df.rename(columns = {"coaster_name" : "Coaster_Name", "year_introduced" : "Year_ Introduced", "latitude" : "Latitude", "longitude": "Longitude", "opening_date_clean":"Opening_Date_Clean", "speed_mph": "Speed_Mph", "height_ft":"Height_Ft", "Inversions_clean":"Inversion","Gforce_clean":"Gforce"})
```

```python
df = df.loc[df.duplicated(subset = [ "Coaster_Name", "Location", "Opening_Date_Clean"])] \
    .reset_index(drop = True).copy()
```

```python
ax = df["Year_ Introduced"].value_counts() \
    .head(15) \
    .plot(kind = "bar", title = "Top 15 Years Coasters Introduced")
ax.set_xlabel("Year Introduced")
ax.set_ylabel("Count")
```
